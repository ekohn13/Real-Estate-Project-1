```python
#Dependencies
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
from pandas.plotting import scatter_matrix

matplotlib.style.use('ggplot')
```

```python
#Opening the files
data_file_population = ("../Desktop/South_Florida_Demographics.csv")
data_file_rents = ("../Desktop/Rents.csv")

data_population_pd = pd.read_csv(data_file_population)
data_rents_pd = pd.read_csv(data_file_rents) 
```

```python
#Cleaining the data and converting to numeric
for col in [i for i in data_population_pd.columns if i not in ['City']]:
    s=data_population_pd[col].apply(lambda x:x.replace(',',''))
    s=s.astype(float)
    data_population_pd[col]=s
```

```python
data_population_pd = data_population_pd.iloc[:-1,:]
data_population_pd
```

```python
total_by_year = []

for row in data_population_pd.iloc[:,1:]:
    s = data_population_pd[row].sum()
    total_by_year.append(s)
```

```python
x = ['Population 2010', 'Population 2011', 'Population 2012', 'Population 2013', 'Population 2014', 'Population 2015', 'Population 2016', 'Population 2017', 'Population 2018']
```

```python
fig,ax = plt.subplots()
plt.plot(total_by_year)
plt.ylabel("Population")
plt.title("Population Growth")
fig.figsize=(40,15)
ax.set_xticklabels(x , rotation=45 )
ax.plot(x,total_by_year,marker='*')
```

```python
total_rents_year = []

for row in data_rents_pd.iloc[:,4:]:
    s = data_rents_pd[row].sum()
    total_rents_year.append(s)

total_year = data_rents_pd.iloc[-1,4:]
```

```python
plt.plot(total_year)
plt.ylabel("Rent prices change")
plt.title("Rents Values")
plt.xlabel("Years")
```
